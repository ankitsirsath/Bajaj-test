```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
df = pd.read_json('/content/DataEngineeringQ2.json')
df.head()
```

```python
with open('/content/DataEngineeringQ2.json') as f:
  d = json.load(f)

df = pd.json_normalize(d)
df.head()
```

```python
df.isna().any()
```

```python
df1=pd.DataFrame(df)
df1.head()
```

```python
print(df1.columns)
```

```python

pD = pd.json_normalize(df1['patientDetails'])
pD.head()
```

```python

cD = pd.json_normalize(df1['consultationData'])
cD.head()
```

```python
# Calculate average number of medicines prescribed
cD['num_medicines'] = cD['medicines'].apply(len)
average_medicines = cD['num_medicines'].mean()

print(f"Average number of medicines prescribed: {average_medicines:.2f}")
```

```python

med = pd.json_normalize(cD['medicines'])
med.head()
```

```python
# Count frequencies of medicine names
medicine_counts = med['medicines'].value_counts()

# Find the 3rd most frequent medicine name
third_most_frequent = medicine_counts.index[2]  # Index 2 corresponds to the 3rd element (0-based index)

print(f"The 3rd most frequently prescribed medicine name is: {third_most_frequent}")
```

```python
# Calculate percentage of missing values for firstName, lastName, and DOB
missing_percentage_firstName = (pD['firstName'].eq('').sum() / len(df)) * 100
missing_percentage_lastName = (pD['lastName'].eq('').sum() / len(df)) * 100
missing_percentage_DOB = (pD['birthDate'].eq('').sum() / len(df)) * 100

# Round off percentages to 2 decimal places
missing_percentage_firstName = round(missing_percentage_firstName, 2)
missing_percentage_lastName = round(missing_percentage_lastName, 2)
missing_percentage_DOB = round(missing_percentage_DOB, 2)

# Print the results
print(f"Percentage of missing values for firstName: {missing_percentage_firstName}%")
print(f"Percentage of missing values for lastName: {missing_percentage_lastName}%")
print(f"Percentage of missing values for DOB: {missing_percentage_DOB}%")
```

```python
# First, impute missing values in 'gender' column with the mode
mode_gender = pD['gender'].mode()[0]  # Calculate the mode of the 'gender' column
pD['gender'].fillna(mode_gender, inplace=True)  # Fill missing values with the mode

# Calculate the percentage of females after imputation
percentage_female = (pD['gender'].value_counts(normalize=True) * 100)['F']

# Print the result
print(f"Percentage of females after imputation: {percentage_female:.2f}%")
```

```python
# Convert 'DOB' to datetime and calculate age
pD['birthDate'] = pd.to_datetime(df['birthDate'])
pD['age'] = (pd.Timestamp.now() - df['birthDate']).astype('<m8[Y]')

# Define function to categorize age into age groups
def categorize_age(age):
    if age <= 12:
        return 'Child'
    elif age <= 19:
        return 'Teen'
    elif age <= 59:
        return 'Adult'
    else:
        return 'Senior'

# Create 'ageGroup' column based on 'age'
pD['ageGroup'] = pD['age'].apply(categorize_age)

# Count of patients in the "Adult" age group
count_adult = pD[pD['ageGroup'] == 'Adult'].shape[0]

print(f"Count of patients in the Adult age group: {count_adult}")
```
